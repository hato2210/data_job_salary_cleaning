The column of the LinkedIn dataframe that we will need to preprocess is salary. The issue here is that some salaries are expressed in terms of British pounds. We will need to convert these into Australia dollars. 

```python
# If a job salary is expressed in terms of British pounds, then convert it into Australian dollars. We also observe that, in such cases, the salary is 
# always an interval and is at an hourly rate. For all other salary value, we leave it as is. 
def pound_to_aud(job):
    salary = job["Salary"]
    if "£" in salary:
        salary_1 = salary.split("-")[0] # Get the lower value of the interval.
        salary_2 = salary.split("-")[1] # Get the upper value of the interval.

        # Get just the number.
        salary_1 = salary_1.strip(" £/hr")
        salary_2 = salary_2.strip(" £/hr")

        # Convert into Australia dollars. We use the approximation of £1 = $2
        salary_1 = str(float(salary_1)*2)
        salary_2 = str(float(salary_2)*2)

        return "$"+salary_1+"/hr -"+" $"+salary_2+"/hr" # Put everything back into the original format. 
    else:
        return salary
```

```python
df_linkedin["Salary"] = df_linkedin.apply(lambda x: pound_to_aud(x), axis=1)
df_linkedin.head()
```

## 4. Preprocess location column

We only consider the states New South Wales, Victoria, and Western Australia. All the other states and territories have a small sample size so we will group them into "Others". 

```python
# We store the states in a dictionary. The keys are the state names in abbreviated form. The value are the keywords that identify a state. 
states = {"NSW": ["nsw", "new south wales"],
          "VIC": ["vic", "victoria"],
          "WA": ["wa", "western australia"],
          "Other": ["qld", "queensland", "sa", "south australia", "tas", "tasmania", "act", "australian capital territory", "nt", "northern territory"]}
# Assign the correct location to a given job.
def get_location(job):
    location = job["location"].lower()
    for state in states: # This loops through the keys. 
        if sum([state_keyword in location for state_keyword in states[state]])>0: # Check if any keywords appear in the location. 
            return state
    return "missing"
```

## 10. Preprocess salary column

```python
# Get the salary of a given job, expressed as a yearly rate. It assumes all salary amounts are prefixed by a dollar sign, and that a rate of pay is 
# included (per hour, per annum, etc...).
def get_salary(job):

    # Get the job salary string in lower case with no whitespaces. Replace all instances of "k" (the abbreviation for $1,000) by "000". 
    salary = job["salary"].replace(" ", "").lower().replace("k", "000")

    # Different rates of pay and their keywords which we use to identify them. Note that we have removed all whitespaces at this point. 
    hourly_keywords = ["anhour", "perhour", "hourly", "/hr", "p.h.", "p/h", "ph"] 
    daily_keywords = ["aday","perday","/day" "daily", "p.d.", "p/d", "pd"]
    weekly_keywords = ["aweek", "weekly", "wk", "p.w.", "p/w", "pw"]
    monthly_keywords = ["amonth", "monthly", "p.m.", "p/m", "pm"]
    yearly_keywords = ["ayear", "peryear", "yearly", "annum", "yr", "p.y.", "p/y", "py", "p.a.", "p/a", "pa"]

    # Check if any keywords appear in the salary string. If so, return the assign the correct multiplier that would convert the amount to a yearly salary.
    if sum([hourly_keyword in salary for hourly_keyword in hourly_keywords]): # If it is an hourly rate of pay.
        multiplier = 1976 # Based on work that is 38 hours per week and 52 weeks a year.
    elif sum([daily_keyword in salary for daily_keyword in daily_keywords]): # If it is a daily rate of pay.
        multiplier = 260 # Based on work that is 5 days a week and 52 week a year. 
    elif sum([weekly_keyword in salary for weekly_keyword in weekly_keywords]): # If it is a weekly rate of pay.
        multiplier = 52 
    elif sum([monthly_keyword in salary for monthly_keyword in monthly_keywords]): # If it is a montly rate of pay.
        multiplier = 12
    elif sum([yearly_keyword in salary for yearly_keyword in yearly_keywords]): # If it is a yearly rate of pay. 
        multiplier = 1
    else: # If there is no rate of pay then we just return "missing".
        return "missing" 
    
    # We determine the number of dollar signs. If there is just one, it means the salary string contains just a single number, for example, this would 
    # occur if the original salary string was "$50 an hour". Here the function would return 98800. If there are two, then it means the salary string 
    # contains a range, for example, this would occur if the original salary string was "$100,000 - $120,000 per annum". Here the function would return 
    # 110000 (the average). However, if there are no dollar signs or there are more than two, then the string probably has no salary information and we 
    # therefore return "missing". 
    num_dollar_signs = sum([char=="$" for char in salary])

    if num_dollar_signs==1:
        # Find the index of "$". 
        for i in range(len(salary)):
            if salary[i]=="$":
                dollar_sign_index = i
                break
        # Having found the "$", keep going until we pass the number.
        past_number = False
        counter = dollar_sign_index
        while not(past_number): 
            counter += 1
            # Check if we have passed the number. If it's not a digit, a decimal point, nor a comma, then we have. Otherwise we have not. 
            past_number = not(salary[counter].isdigit() or salary[counter]=="." or salary[counter]==",") 
        salary_cleaned = salary[dollar_sign_index+1:counter] # Get the number (dollar sign excluded) within the salary string.
        return float(salary_cleaned.replace(",", ""))*multiplier # Convert it to a yearly salary. 
    
    elif num_dollar_signs==2:
        # Find the index of the first instance of "$".
        for i in range(len(salary)):
            if salary[i]=="$":
                dollar_sign_first_index = i
                break
        # Having found the first "$", keep going until we pass the number associated to this "$". 
        past_number = False
        counter = dollar_sign_first_index
        while not(past_number):
            counter += 1
            past_number = not(salary[counter].isdigit() or salary[counter]=="." or salary[counter]==",")
        salary_lower = salary[dollar_sign_first_index+1:counter].replace(",", "") # Store the lower limit of the salary range.
        # Find the index of the second (and final) instance of "$". 
        for i in range(counter+1, len(salary)):
            if salary[i]=="$":
                dollar_sign_second_index = i
                break
        # Having found the second "$", keep going until we pass the number associated to this "$". 
        past_number = False
        counter = dollar_sign_second_index
        while not(past_number): 
            counter += 1
            try:
                past_number = not(salary[counter].isdigit() or salary[counter]=="." or salary[counter]==",")
            except IndexError: 
                # This occurs when there are no text beyond the upper salary. An example of such a value from the dataframe is
                # "$750 per day including super / $165k". Such values complicate things since we will have to determine whether it is a single salary or
                # an interval that we are dealing with. Here we take the simpler option and just return "missing".
                return "missing"
        salary_upper = salary[dollar_sign_second_index+1:counter].replace(",", "") # Stroe the upper limit of the salary range.

        return (float(salary_lower)+float(salary_upper))/2*multiplier # Take the average of the lower and upper limit then convert it to a yearly salary.

    else:
        return "missing"
```

## 11. Add rate of pay column 

The rate of pay of a salary could be hourly, daily, weekly, monthly, or yearly. In this section we add five boolean columns, one for each rate of pay. 

```python
# Returns "True" if the job salary is paid hourly, "False" otherwise. 
def is_hourly(job):
    hourly_keywords = ["anhour", "perhour", "hourly", "/hr", "p.h.", "p/h", "ph"] # Keywords that suggests hourly pay.
    not_hourly_keywords = ["aday","perday","/day" "daily", "p.d.", "p/d", "pd", "aweek", "weekly", "wk", "p.w.", "p/w", "pw", "amonth", "monthly", 
                           "p.m.", "p/m", "pm", "ayear", "peryear", "yearly", "annum", "yr", "p.y.", "p/y", "py", "p.a.", "p/a", "pa"] # Keywords that suggest another rate of pay. 
    salary = job["salary"].lower()
    if sum([hourly_keyword in salary for hourly_keyword in hourly_keywords])>0: 
        return 1
    elif sum([not_hourly_keyword in salary for not_hourly_keyword in not_hourly_keywords])>0:
        return 0
    else: # Occurs when the salary string has no rate of pay. In such cases we assume that it is not an actual salary, and so we return "missing".
        return "missing"
```

```python
# Returns "True" if the job salary is paid daily, "False" otherwise. 
def is_daily(job):
    daily_keywords = ["aday","perday","/day" "daily", "p.d.", "p/d", "pd"]
    not_daily_keywords = ["anhour", "perhour", "hourly", "/hr", "p.h.", "p/h", "ph", "aweek", "weekly", "wk", "p.w.", "p/w", "pw", "amonth", "monthly", 
                          "p.m.", "p/m", "pm", "ayear", "peryear", "yearly", "annum", "yr", "p.y.", "p/y", "py", "p.a.", "p/a", "pa"]
    salary = job["salary"].lower()
    if sum([daily_keyword in salary for daily_keyword in daily_keywords])>0:
        return 1
    elif sum([not_daily_keyword in salary for not_daily_keyword in not_daily_keywords])>0:
        return 0
    else:
        return "missing"
```

```python
# Returns "True" if the job salary is paid weekly, "False" otherwise. 
def is_weekly(job):
    weekly_keywords = ["aweek", "weekly", "wk", "p.w.", "p/w", "pw"]
    not_weekly_keywords = ["anhour", "perhour", "hourly", "/hr", "p.h.", "p/h", "ph", "aday","perday","/day" "daily", "p.d.", "p/d", "pd", "amonth", 
                           "monthly", "p.m.", "p/m", "pm", "ayear", "peryear", "yearly", "annum", "yr", "p.y.", "p/y", "py", "p.a.", "p/a", "pa"]
    salary = job["salary"].lower()
    if sum([weekly_keyword in salary for weekly_keyword in weekly_keywords])>0:
        return 1
    elif sum([not_weekly_keyword in salary for not_weekly_keyword in not_weekly_keywords])>0:
        return 0
    else:
         return "missing"
```

```python
# Returns "True" if the job salary is paid yearly, "False" otherwise. 
def is_yearly(job):
    yearly_keywords = ["ayear", "peryear", "yearly", "annum", "yr", "p.y.", "p/y", "py", "p.a.", "p/a", "pa"]
    not_yearly_keywords = ["anhour", "perhour", "hourly", "/hr", "p.h.", "p/h", "ph", "aday","perday","/day" "daily", "p.d.", "p/d", "pd", "aweek", 
                           "weekly", "wk", "p.w.", "p/w", "pw", "amonth", "monthly", "p.m.", "p/m", "pm"]
    salary = job["salary"].lower()
    if sum([yearly_keyword in salary for yearly_keyword in yearly_keywords])>0:
        return 1
    elif sum([not_yearly_keyword in salary for not_yearly_keyword in not_yearly_keywords])>0:
        return 0
    else:
        return "missing"
```

```python
df["hourly"] = df.apply(lambda x: is_hourly(x), axis=1)
df["daily"] = df.apply(lambda x: is_daily(x), axis=1)
df["weekly"] = df.apply(lambda x: is_weekly(x), axis=1)
df["monthly"] = df.apply(lambda x: is_monthly(x), axis=1)
df["yearly"] = df.apply(lambda x: is_yearly(x), axis=1)
df.head()
```

## 12. Add Job Level Column.

We consider the following job levels:
1. Non-senior
2. Senior

Alternative we could have classified a job level into junior, mid, or senior. However, due to the problems with small sample sizes we have decided to pool the junior and mid levels into non-senior. 

```python
senior_keywords = ["senior", "principal", "head", "manager", "lead"]
def get_job_level(job):
    job_title = job["job_title"].lower()
    if sum([senior_keyword in job_title for senior_keyword in senior_keywords])>0:
        return 1
    else:
        return 0
```

```python
df["senior"] = df.apply(lambda x: get_job_level(x), axis=1)
```
